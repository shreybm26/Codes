```python
df1=pd.DataFrame({"name":['Shrey', 'Shreya', 'Hello'],
                 "toy":[np.nan, 'Genshin', 'Handshake'],
                 "born": [pd.NaT, pd.Timestamp("2002-12-17"), pd.NaT]})
df1.head()
```

```python
df1.dropna()
```

```python
df1.dropna(how="all")

#Drop the null value only if all the data is null
```

```python
df2.drop_duplicates(subset=['name'])
```

```python
df2.drop_duplicates(subset=['name'], keep='last')
```
