```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

BASIC GRAPHS

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#Resizing the graph
plt.figure(figsize=(4,2))
#plt.plot(x,y, label='x^2', color='red', linestyle='dashed', linewidth=1.5, marker='o')

#Using shorthand notations
#fmt='[color][marker][line]'

plt.plot(x,y, 'bo--', label='2x')
plt.legend()

x2=np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2, 'r-', label='x^2')
plt.plot(x2[5:], x2[5:]**2, 'r--', label='x^2')

plt.title('Square of numbers', fontdict={'fontname': 'JetBrains Mono', 'fontsize': 10})
plt.xlabel('n', fontdict={'fontname': 'Cambria'})
plt.ylabel('f(n)', fontdict={'fontname': 'Cambria'})

plt.xticks([1,2,3,4])
#plt.yticks([2,4,6,8])

plt.savefig('mygraph.png', dpi=300)

plt.show()
```

BAR GRAPHS

```python
labels=['A', 'B', 'C']
values=[2,4,9]

plt.figure(figsize=(4,2))

bars=plt.bar(labels, values)
patterns=['/', '+', '*']

for i in bars:
    i.set_hatch(patterns.pop(0))

#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('*')

plt.show()
```

PRACTICING WITH REAL-WORLD DATA

```python
gas=pd.read_csv('gas_prices.csv')
plt.figure(figsize=(5,3))

plt.title('Gas Prices Worldwide', fontdict={'fontweight': 'bold', 'fontsize': 12, 'fontname':'Cambria'})

plt.plot(gas.Year, gas.USA, 'b-', label='USA')
plt.plot(gas.Year, gas.Japan, 'r-', label='Japan')
plt.plot(gas.Year, gas.Germany, 'y-', label='Germany')

countries_to_put=['Mexico', 'France', 'Italy']

for country in gas:
    if country in countries_to_put:
        plt.plot(gas.Year, gas[country], marker='.')

plt.xlabel('Year')
plt.ylabel('USD')

plt.xticks(gas.Year[::3])

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```

FIFA DATA

```python
fifa=pd.read_csv('fifa_data.csv')
fifa.head()
```

```python
bins=[]
for i in range(40,110,10):
    bins.append(i)
    i=i+10

plt.hist(fifa.Overall, bins=bins, color='#abcdef')

plt.ylabel='Number of Players'
plt.xlabel='Skill Level'
plt.title='Performance Stats of Players'

plt.xticks(bins)

plt.show()
```

PIE CHARTS

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

labels=['Left', 'Right']
colors=['#acbdef', "#ebcdfa"]
explode=(.1,.1)
plt.pie([left,right], labels=labels, colors=colors, autopct='%.2f %%', explode=explode)
plt.show()
```
